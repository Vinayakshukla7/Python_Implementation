```python
import re                                                        #importing libraries
import pdfplumber
import pandas as pd
from collections import namedtuple
```

```python
Line = namedtuple('Line', 'Gastname Total Netto_EUR MwSt_EUR Brutto_EUR')  #namedtuple used for making tuple class 
```

```python
gastname_re = re.compile(r'^Gastname :(\w+ .*)')                        #using regex to locate the data from rows
line_re = re.compile(r'(\d*\.\d{2}) (\d*\.\d{2}) (\d*\.\d{2})')
total_re=re.compile(r'(\d{3}\.\d{2})')
```

```python
file = 'Multi_lines.pdf'                                                           #importing pdf 
```

```python
#from opening pdf to append the data in tuples .
lines = []                                                      
with pdfplumber.open(file) as pdf:                          
    pages = pdf.pages
    for page in pdf.pages:
        print(page)
        text = page.extract_text()
        print(text)
        for line in text.split('\n'):
            comp = gastname_re.search(line)
            if comp:
                Gastname = comp.group(1)
            elif line.startswith('Umsatzsteuer'):
                Total = total_re.search(line).group(1)
            elif line.startswith('Total'):
                sel=line_re.search(line)
                Netto_EUR=sel.group(1)
                MwSt_EUR=sel.group(2)
                Brutto_EUR=sel.group(3)
                lines.append(Line(Gastname,Total,Netto_EUR,MwSt_EUR,Brutto_EUR))
                
```

```python
df = pd.DataFrame(lines)                
```

```python
df.to_csv('Multi_Pages.csv')
```
