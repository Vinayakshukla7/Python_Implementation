# What is tabula ?
In simple words ,Tabula is a tool for liberating data tables locked inside PDF files.<br>
tabula-py is a simple Python wrapper of tabula-java, which can read tables in a PDF. You can read tables from a PDF and convert them into a pandas DataFrame. tabula-py also enables you to convert a PDF file into a CSV, a TSV or a JSON file.<br>

tabula-py is good for:<br>

automation with Python script<br>
advanced analytics after converting pandas DataFrame<br>
casual analytics with Jupyter notebook or Google Colabolatory<br>

## .read_pdf
 tabula-py can load a PDF or file like object on both local or internet by using read_pdf() function.

```python
pdf_path = "https://github.com/chezou/tabula-py/raw/master/tests/resources/data.pdf"

dfs = tabula.read_pdf(pdf_path, stream=True)
# read_pdf returns list of DataFrames
print(len(dfs))
dfs[0]

#If you want to extract all pages, add ``pages="all"`` in argument of .read_pdf.
```

### Set page option

```python
dfs = tabula.read_pdf(pdf_path, pages=2, stream=True)                   # only 2nd page                 
dfs[0]
```

```python
dfs = tabula.read_pdf(pdf_path, pages='all', stream=True)           # all pages are selected                  
dfs        
```

## Read partial area of PDF
If you want to set a certain part of page, you can use area option.

Note: tabula-py 2.0.0, multiple_tables option became True so if you want to use multiple area options like [[0, 0, 100, 50], [0, 50, 100, 100]], you need to set multiple_tables=False.

area=[top,left,down,right]

```python
dfk = tabula.read_pdf(pdf_path, area=[126,149,212,462], pages=2)
dfk[0]
```

### Read giving column information

```python
pdf_path2 = "https://github.com/chezou/tabula-py/raw/master/tests/resources/campaign_donors.pdf"

dfs = tabula.read_pdf(pdf_path2, columns=[47, 147, 256, 310, 375, 431, 504], guess=False, pages=1)
df = dfs[0].drop(["Unnamed: 0"], axis=1)
df
```

## Convert PDF tables into CSV, TSV, or JSON files

```python
tabula.convert_into(pdf_path2, "test.csv", output_format="csv",pages="all")
```
