```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup
import requests
import pandas as pd  
```

```python
def get_soup(url):
    s = HTMLSession() 
    info=s.get(url)
    soup=BeautifulSoup(info.text,'html.parser')
    return soup
```

```python
lists=[]
#to_links=[]
soups=get_soup("https://www.chittorgarh.com/report/ipo-in-india-list-main-board-sme/82/")
all_rows=soups.find_all('tbody')
#print(all_rows)
for one in all_rows:
    x=one.find_all('tr')
    for two in x:
        main_dict={'Issuer_Company':'','Exchange':'','Open':'','Close':'','Lot_size':'','issue_price':'','Issue_size':'','Face_value':'','Fresh_issue':'','Offer_for_scale':'','init_refund':'','ipo_listing':'','Qualified_Institutional':'','Non_Institutional':'','Retail_Individual':'','Employee':''}
        universal_dict={'0':'Issuer_Company','1':'Exchange','2':'Open','3':'Close','4':'Lot_size','5':'issue_price','6':'Issue_size','7':'Face_value','8':'Fresh_issue','9':'Offer_for_scale','10':'init_refund','11':'ipo_listing','12':'Qualified_Institutional','13':'Non_Institutional','14':'Retail_Individual','15':'Employee'}
        link=two.find('a')['href']
        every_link(link)
        #to_links.append(link)
        y=two.find_all('td')
        for i,three in enumerate(y):
            value=universal_dict[f'{i}']
            main_dict[f'{value}']=f'{three.text}'
            #print(f"{i} has {three.text}")
        #print("________________________________")
        lists.append(main_dict)
```
