```python
import pandas as pd
import re
import numpy as np
import pygsheets
```

```python
da=da.fillna(-999)
```

```python
da['Open']=pd.to_datetime(da.Open)                                #cleaning of dates
da['Close']=pd.to_datetime(da.Close)
da['init_refund']=pd.to_datetime(da.init_refund)
da['ipo_listing']=pd.to_datetime(da.ipo_listing)
```

```python
da=da.replace(['[.]','[.] Times','['],['-999','-999','-999'])
```

```python
da['Issue_size']=da['Issue_size'].round(0).astype(int)
da['Lot_size']=da['Lot_size'].round(0).astype(int)
da
```

```python
#Pandas has a built in replace method for "object" columns.
da['Employee'] = da['Employee'].str.replace(",","").astype(float)
da['Employee'] = da['Employee'].fillna(-999)
da['Employee']=da['Employee'].round(0).astype(int)
```

```python
da['Retail_Individual'] = da['Retail_Individual'].astype(float)
da['Retail_Individual'] = da['Retail_Individual'].fillna(-999)
da['Retail_Individual']=da['Retail_Individual'].round(0).astype(int)
```

```python
da['Non_Institutional']=da['Non_Institutional'].round(0).astype(int)
```

```python
da['Qualified_Institutional'] = da['Qualified_Institutional'].astype(float)
da['Qualified_Institutional'] = da['Qualified_Institutional'].fillna(-999)
da['Qualified_Institutional']=da['Qualified_Institutional'].round(0).astype(int)
```

```python
def Issue_price(name):                                                    #cleaning of Issue price
    # Search for opening bracket in the name followed by
    # any characters repeated any number of times
    if re.search('[to]', str(name)):
  
        # Extract the position of beginning of pattern
        pos = re.search('[to]', str(name)).start()
        pos = pos+3
        # return the cleaned name
        name = str(name)[pos:]
        return int(name)
  
    else:
        # if clean up needed return the same name
        return int(name)
          
# Updated the city columns
da['issue_price'] = da['issue_price'].apply(Issue_price)
```

```python
def face_value(text):            #cleaning face value
    text=text.split(' ')
    text=text[0].split('₹')
    return int(text[1])

          
# Updated the city columns
da['Face_value'] = da['Face_value'].apply(face_value)
```

```python
da['Fresh_issue']=da['Fresh_issue'].fillna(-999)
for x in range(len(da['Fresh_issue'])):
    val=str(da['Fresh_issue'][x])
    if val != '-999':
        val=str(val)
        text=val.split('up to')
        text=text[1].split(' ')
        text=text[1].split('₹')
        text=text[1].split('.')
        text=text[0]
        da['Fresh_issue'][x]=str(text)
```

```python
da['Offer_for_sale']=da['Offer_for_sale'].fillna(-999)
for x in range(len(da['Offer_for_sale'])):
    val=str(da['Offer_for_sale'][x])
    if val != '-999':
        val=str(val)
        text=val.split('up to')
        text=text[1].split(' ')
        text=text[1].split('₹')
        text=text[1].split('.')
        text=text[0]
        da['Offer_for_sale'][x]=str(text)
```
