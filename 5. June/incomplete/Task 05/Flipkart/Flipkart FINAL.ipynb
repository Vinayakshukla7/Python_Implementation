```python
from requests_html import HTMLSession
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
links=[]
```

```python
def get_soup(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup

def page_number(base_url):
    soup=get_soup(base_url)
    no=soup.find('div',{'class':'_2MImiq'})
    pg_no=no.find('span').text.split(' ')[3]
    return pg_no

def links_to_review(url):
    sess = HTMLSession() 
    info = sess.get(url)
    soup=BeautifulSoup(info.text,'html.parser')
    above = soup.find_all('div',{'class':'_2kHMtA'})
    for x in above:
        try:
            y=x.find('a',{'class':'_1fQZEK'})['href']
            z=y.split('&q')[0]
            z='https://www.flipkart.com'+str(z)+str("&page=")
            final_link=z.replace('/p/','/product-reviews/')
            links.append(final_link)
        except:
            pass
```

```python
search_by='samsung smartphone'
initial_page=1
ini_url=f'https://www.flipkart.com/search?q={search_by}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={initial_page}'

total_page=page_number(ini_url)
for current in range(1,int(total_page)+1):
    url_new=f'https://www.flipkart.com/search?q={search_by}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={current}'
    links_to_review(url_new)
```

```python
reviewlist = []
```

```python
def get_soup(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup

def reviews_pgno(base_url):
    soup=get_soup(base_url)
    no=soup.find('div',{'class':'_2MImiq _1Qnn1K'})
    if no is not None:
        pg_no=no.find('span').text.split(' ')[3]
        return pg_no
    else:
        return 

def all_reviews(soup):
    reviews = soup.find_all('div', {'class': '_27M-vq'})
    try:
        for x in reviews:
            reviews={
            'Phone_name':soup.find('div', {'class': '_2s4DIt _1CDdy2'}).text.replace('Reviews',''),
            'Review_title':x.find('div', {'class': 'col _2wzgFH K0kLPL'}).find('p').text,
            'Review_body':x.find('div',{'class':''}).text.replace('READ MORE',''),
            'Ratings':x.find('div',{'class':'_3LWZlK _1BLPMq'}).text.strip()                         #strip avoids wrong results here
            }
            reviewlist.append(reviews)
    except:
        pass      

#base_url='https://www.flipkart.com/redmi-9-sporty-orange-64-gb/product-reviews/itm4fb151383983b?pid=MOBFV5FPCJC9ZKRB&lid=LSTMOBFV5FPCJC9ZKRBIAYSS8&marketplace=FLIPKART&page=1'
y=1
for product in links:
    total_pages=reviews_pgno(product)
    if total_pages is not None:
        total_pages=total_pages.replace(',', '')
        for x in range(1,int(total_pages)+1):
            new_urls=f'{product}{x}'
            print(new_urls)
            soup=get_soup(new_urls)
            all_reviews(soup)
            if x ==10:                                                          #decides how many pages
                break
    if y==10:                                                                     #decides how many products
        break
    y=y+1
```
