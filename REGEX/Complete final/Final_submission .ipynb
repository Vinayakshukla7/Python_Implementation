```python
from selenium import webdriver  
import pandas as pd
from selenium.webdriver.support.ui import Select     
from selenium.webdriver.common.keys import Keys
import time
```

```python
driver = webdriver.Chrome("D:\Sem 6\Selenium\chromedriver.exe")
driver.get("http://mpsecerms.mp.gov.in/SECSearchEngine")  
show=driver.find_element_by_xpath("/html/body/div[2]/div/div/div/ul/li[2]/a")       #get through the right section t
show.click() 
sel=Select(driver.find_element_by_xpath("//*[@id='DISTNO']"))

sel.select_by_index(22) 

excel_file = r'C:\Users\vsvin\aa.xlsx'                                   #The excel file with inputs 
numbers=pd.read_excel(excel_file, header=None)
mylist=list(numbers[0])

"""
mylist = ['SGU1710524',
'SGU2126076',
'MP/37/270/423153',
'BVC2684108',
'BVC2684181',
'SGU6970719',
'SGU6970727',
'BVC2379022',
'SGU6970396',
'BVC2379188',
'SGU7010671',
'SGU6901912',
'SGU2368579',
'SGU2230134',
'SGU2479749',
'SGU2479756',
'SGU2524064',
'SGU2524106',
'SGU2524114',
'SGU2528453',
'SGU2528479',
'SGU2528511',
'SGU2528594',
'SGU2555100',
'SGU2586006'
]

"""
for i in range(len(mylist)):
    try:
        show=driver.find_element_by_xpath("//*[@id='EpicNo']")
        show.clear()
        show.send_keys(mylist[i])
        show=driver.find_element_by_xpath("//*[@id='btnSearchByEpic']")
        show.send_keys(Keys.RETURN)
        time.sleep(3)
        rows = len(driver.find_elements_by_xpath("//*[@id='forms_wrapper']/div/div"))
        cols = len(driver.find_elements_by_xpath("//*[@id='forms']/tbody/tr/td"))
        row_names=[]
        time.sleep(3) 
        if i==0:
            col_names=[]
            for p in range(1, cols-3):
                value = driver.find_element_by_xpath("//*[@id='forms_wrapper']/div/div[1]/div/table/thead/tr/th["+str(p)+"]").text
                col_names.append(value)
                value = driver.find_element_by_xpath("/html/body/div[2]/div/div/div/div[3]/div[3]/div/div/div[2]/div/div/div/div[2]/table/tbody/tr/td["+str(p)+"]").text
                row_names.append(value)

            DKS=pd.DataFrame(columns=col_names)

        else:
            for p in range(1, cols-3):
                value = driver.find_element_by_xpath("/html/body/div[2]/div/div/div/div[3]/div[3]/div/div/div[2]/div/div/div/div[2]/table/tbody/tr/td["+str(p)+"]").text
                row_names.append(value)
            time.sleep(3)  
            driver.find_element_by_xpath("/html/body/div[2]/div/div/div/div[3]/div[3]/div/div/div[1]/button").click()
        DKS.loc[len(DKS)] = row_names
        time.sleep(2)
    except ValueError:
        print(f"{mylist[i]} has no data")
        continue
    
driver.quit()
```
